```python
sns.pairplot(df[['Price', 'Market Cap', 'Volume (24h)', '24h', '7d']], diag_kind='kde')
plt.suptitle('Pairplot of Key Metrics', y=1.02)
plt.show()
```

```python
top_10 = df.nlargest(10, 'Volume (24h)')

plt.figure(figsize=(10, 6))
sns.barplot(x='Volume (24h)', y='Name', data=top_10, palette='coolwarm')
plt.title('Top 10 Cryptocurrencies by Volume')
plt.xlabel('Volume (24h)')
plt.ylabel('Cryptocurrency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True, color='green', log_scale=True)
plt.title('Distribution of Prices (Log Scale)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Market Cap', y='Volume (24h)', hue='7d', size='Price', sizes=(50, 500), data=df, alpha=0.7)
plt.title('Market Cap vs. Volume (Hue by 7d Fluctuation, Size by Price)')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Market Cap (log)')
plt.ylabel('Volume (log)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='24h', data=df, palette='pastel')
plt.title('24h Fluctuation Distribution')
plt.xlabel('24h Fluctuation (%)')
plt.show()
```
